```python
import osmnx as ox
%matplotlib inline
ox.config(log_console=True, use_cache=True)
ox.__version__
```

```python
gdf = ox.graph_from_bbox(-37.705, -37.796, 144.949, 144.84, network_type='walk')
gdf = ox.project_graph(gdf)
ox.plot_graph(gdf)
```

```python
# save street network as ESRI shapefile
ox.save_graph_shapefile(gdf, folder = 'shp/out', filename='network-shape')
```
